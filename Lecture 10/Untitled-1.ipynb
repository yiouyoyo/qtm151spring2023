```python
import math
import time

# Define the height and width of the wave
height = 20
width = 60

# Define the frequency and speed of the wave
freq = 4
speed = 0.5

# Loop through each row of the screen
while True:
    # Clear the screen
    print('\033[2J', end='')
    
    # Calculate the current time
    t = time.time() * speed
    
    # Loop through each column of the screen
    for x in range(width):
        # Calculate the height of the wave at this point
        y = int((height/2) * math.sin(2*math.pi*freq*((x/width)-t)) + (height/2))
        
        # Print the character at this position
        if y == height:
            print('_', end='')
        elif y == 0:
            print('_', end='')
        elif y < height/2:
            print('/', end='')
        else:
            print('\\', end='')
            
    # Move to the next row
    print()
    time.sleep(0.05)


```
